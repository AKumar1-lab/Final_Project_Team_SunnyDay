```python
pv=pd.read_csv('pv_open_2020.csv')
pv
```

```python
missing=zips.loc[zips['ghi']=='none']
missing
```

```python
for i in range(len(missing)):
    value=pv.loc[(pv['latitude']>=missing['latitude'].iloc[i]-.2)&(pv['latitude']<=missing['latitude'].iloc[i]+.2)&(pv['longitude']>=missing['longitude'].iloc[i]-.2)&(pv['longitude']<=missing['longitude'].iloc[i]+.2)]['global_horizontal_irradiance'].mean()
    missing['ghi'].iloc[i]=value
```

```python
missing=missing.fillna('none')
missing
```

```python
ghi=pd.DataFrame(missing[['zip','ghi']])
ghi
```

```python
ghi.to_csv('new_values.csv')
```
