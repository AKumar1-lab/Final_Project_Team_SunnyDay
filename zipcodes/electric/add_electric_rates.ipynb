```python
import pandas as pd
```

```python
# Residential electric rates by state
elecrates=pd.read_csv('2020avg.csv')
elecrates
```

```python
# Zipcodes with GHI
zips=pd.read_csv('cleaned_zips.csv')
zips=zips.drop(['Unnamed: 0'],axis=1)
zips
```

```python
# Add electric rates to zipcode dataframe
zips=pd.merge(left=zips,right=elecrates,left_on='state',right_on='State',how='left')
```

```python
# Drop duplicate columns
zips=zips.drop(['Year','State'],axis=1)
```

```python
# Add units
zips=zips.rename(columns={'ghi':'global_horizontal_irradiance_(kWh/m^2)/day','Residential':'residential_electric_rate_(cents/kWh)'})
```

```python
zips.to_csv('zipcodes_ghi_electric.csv')
```
